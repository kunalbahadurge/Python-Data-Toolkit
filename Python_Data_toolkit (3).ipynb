# **Data Toolkit theory questions**

    1. What is NumPy, and why is it widely used in Python
    = NumPy is a Python library used for numerical and scientific computing. It provides powerful tools to work with arrays, matrices, and large numerical datasets efficiently. It is widely used because NumPy arrays are faster and more memory-efficient than Python lists, as they store data in a continuous block of memory.


    2. How does broadcasting work in NumPy
    = Broadcasting allows NumPy to perform operations on arrays of different shapes without writing loops. It automatically adjusts smaller arrays to match larger ones. The smaller array is virtually expanded so that element-wise operations can be performed easily and correctly.


    3. What is a Pandas DataFrame
    = A Pandas DataFrame is a two-dimensional data structure that stores data in rows and columns, similar to a table or spreadsheet. It can store different types of data such as numbers, strings, and dates in different columns. Each column has a label, and each row has an index, making data selection easy.


    4. Explain the use of the groupby() method in Pandas
    = The groupby() method is used to divide data into groups based on one or more columns. After grouping, operations like sum, mean, count, and max can be applied to each group. It is useful for analyzing category-wise data, such as sales per region or marks per class.


    5. Why is Seaborn preferred for statistical visualizations
    = Seaborn provides high-level functions for creating statistical graphs easily.It automatically handles aesthetics like colors, labels, and themes.


    6. What are the differences between NumPy arrays and Python lists
    = NumPy arrays store elements of the same data type, while Python lists can store mixed types. Arrays are faster than lists because they are stored in continuous memory. NumPy supports vectorized operations, whereas lists require loops. Arrays consume less memory compared to lists.


    7. What is a heatmap, and when should it be used
    = A heatmap is a graphical representation of data using colors to show values. Darker or brighter colors represent higher or lower values. Heatmaps are commonly used to show correlations between variables. They help quickly identify patterns and relationships in data.


    8. What does the term “vectorized operation” mean in NumPy
    = Vectorized operations apply calculations to entire arrays at once. They eliminate the need for explicit loops in Python. Operations are faster because they run in optimized C code. Vectorization makes code cleaner and easier to read.


    9. How does Matplotlib differ from Plotly
    = Matplotlib creates static plots that are mainly used in reports and notebooks. Plotly creates interactive graphs with zooming, hovering, and clicking features. Plotly is better suited for web-based dashboards. Matplotlib is simpler and more commonly used for basic plots.


    10. What is the significance of hierarchical indexing in Pandas
    = Hierarchical indexing allows multiple index levels in a DataFrame. It helps represent higher-dimensional data in a 2D format. Data selection becomes more flexible using multiple indexes. It is useful when working with grouped or structured datasets.



    11. What is the role of Seaborn’s pairplot() function
    = pairplot() displays relationships between multiple variables at once. It shows scatter plots and histograms together. Useful for exploratory data analysis to detect trends. Helps identify correlations and patterns quickly.


    12. What is the purpose of the describe() function in Pandas
    = describe() provides a statistical summary of numerical data. It shows values like mean, count, minimum, maximum, and quartiles. Helps understand data distribution and spread. Useful for quick data inspection.



    13. Why is handling missing data important in Pandas
    = Missing data can lead to incorrect analysis and results. Pandas provides methods to remove or fill missing values. Proper handling improves data accuracy and reliability. Clean data leads to better decisions and models.


    14. What are the benefits of using Plotly for data visualization
    = Plotly provides interactive and dynamic visualizations. Users can zoom, hover, and explore data easily. Suitable for dashboards and web applications. Enhances user engagement and understanding.


    15. How does NumPy handle multidimensional arrays
    = NumPy supports arrays with multiple dimensions such as 2D and 3D arrays. It uses shape and axis concepts to manage data efficiently. Allows matrix and tensor operations easily. Efficient memory usage improves performance.


    16. What is the role of Bokeh in data visualization
    = Bokeh is used to create interactive visualizations for web browsers. It supports real-time data updates. Suitable for large and complex datasets. Commonly used in dashboards and web apps.


    17. Explain the difference between apply() and map() in Pandas
    = map() works only on Pandas Series. apply() works on both Series and DataFrames. apply() is more flexible for complex operations. map() is faster for simple element-wise operations.


    18. What are some advanced features of NumPy
    = Broadcasting and vectorized operations. Linear algebra and matrix operations. Random number generation tools. Integration with other scientific libraries.


    19. How does Pandas simplify time series analysis
    = Pandas supports date-time indexing. It allows easy resampling and shifting of data. Handles missing time values efficiently. Useful for financial and trend analysis.


    20. What is the role of a pivot table in Pandas
    = Pivot tables summarize large datasets. They perform aggregation like sum and mean. Help rearrange data for better analysis. Widely used in reports and analytics.


    21. Why is NumPy’s array slicing faster than Python’s list slicing
    = NumPy arrays use contiguous memory blocks. Operations are handled in optimized C code. Avoids Python-level loops. Results in faster execution.



    22.  What are some common use cases for Seaborn
    = Exploratory data analysis. Statistical data visualization. Correlation and distribution analysis. Relationship and comparison plots.

# **Practical questions**

1. How do you create a 2D NumPy array and calculate the sum of each row

```python
import numpy as np

arr = np.array([[2, 4, 6], [1, 3, 5]])
row_sum = arr.sum(axis=1)
col_sum = arr.sum(axis=0)

print(row_sum)
print(col_sum)
```

2. Write a Pandas script to find the mean of a specific column in a DataFrame

```python
import pandas as pd

df = pd.DataFrame({'Scores': [65, 70, 85, 90, 75]})
mean_val = df['Scores'].mean()
median_val = df['Scores'].median()
std_val = df['Scores'].std()

print(mean_val, median_val, std_val)
```

3. Create a scatter plot using Matplotlib

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y)
plt.show()
```

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {'maths': [1, 2, 3], 'science': [4, 5, 6], 'english': [7, 8, 9]}
df = pd.DataFrame(data)

corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

5.  Generate a bar plot using Plotly

```python
import plotly.express as px
import pandas as pd

data = pd.DataFrame({
    'Product': ['A', 'B', 'C', 'D'],
    'Sales': [120, 250, 180, 300]
})

fig = px.bar(data, x='Product', y='Sales', title='Product Sales')
fig.show()
```

11. Perform matrix multiplication using NumPy

```python
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

result = np.matmul(a, b)
print(result)
```

13.  Create a 3D scatter plot using Plotly

```python
import plotly.express as px
import plotly.io as pio
import pandas as pd

data = pd.DataFrame({
    'X': [1, 2, 3, 4, 5],
    'Y': [5, 4, 3, 2, 1],
    'Z': [10, 20, 15, 25, 30]
})

fig = px.scatter_3d(
    data,
    x='X',
    y='Y',
    z='Z',
    title='3D Scatter Plot Example'
)

fig.show('colab')
```
